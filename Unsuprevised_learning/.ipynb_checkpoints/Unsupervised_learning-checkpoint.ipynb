## Hierachical clustering 

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn import metrics
```

```python
path = './in/'
group = '02' 
ds1 = 1 # dataset 1
ds2 = 2 # dataset 2
data1 = np.loadtxt(path + 'ds'+group+str(ds1)+'.txt')
data2 = np.loadtxt(path + 'ds'+group+str(ds2)+'.txt')

X1 = data1[:, 0:2]
y1 = data1[:, 2:3].ravel()

X2 = data2[:, 0:2]
y2 = data2[:, 2:3].ravel()
```

Se define esta función que permitirá realizar más fácilmente el ejercicio 1.

```python
def hierarchichal_clustering(method, n_clusters, ds1, ds2, labels1, labels2):
    
    print ('Method: ', method)
    print ('Number of clusters: ', n_clusters)
    #se instancia el método demandado
    clust = AgglomerativeClustering(n_clusters=n_clusters, linkage=method) 
    #se predicen las clases de los dos conjuntos de datos
    pred_1 = clust.fit_predict(ds1) 
    pred_2 = clust.fit_predict(ds2)
    # V-measure
    v1 = metrics.v_measure_score(labels1, pred_1) #grupo 1
    v2 = metrics.v_measure_score(labels2, pred_2) #grupo 2
    #Davies-Bouldin
    db1 = metrics.davies_bouldin_score(ds1, pred_1) #grupo 1
    db2 = metrics.davies_bouldin_score(ds2, pred_2) #grupo 2
    
    #Se retornan los valores de las métricas
    return v1, v2, db1, db2
```

## Cost function optimization-based clustering

### Crisp and fuzzy clustering

```python
import math
```

`distance` calcula la distancia de un punto a una recta

```python
def distance (x, rad):
    
    # x = punto 
    # rad = parámetro del cluster (ángulo de la recta en radianes)
    
    return (np.sin(rad)**2)*x[0]**2 + (np.cos(rad)**2)*x[1]**2 - 2*np.sin(rad)*np.cos(rad)*x[0]*x[1]
```

`distance_array` realiza lo mismo que `distance` pero con un conjunto o array de puntos

```python
def distance_array (x_arr, rad):
    
    # x_arr = conjunto de puntos
    # rad = parámetro del cluster (ángulo de la recta en radianes)
    
    return (np.sin(rad)**2)*np.square(x_arr[:,0]) + (np.cos(rad)**2)*np.square(x_arr[:,1]) - 2*np.sin(rad)*np.cos(rad)*x_arr[:,0]*x_arr[:,1]
```

`theta_clust` determina los parámetros $\theta$. Corresponde con la derivada parcial de la función de coste respecto dichos parámetros.

```python
def theta_clust (x, ux, q = 1): #q = 1 for crisp clustering
    
    # x = conjunto de puntos 
    # ux = pertenencias
    
    sum1 = np.sum(np.multiply(np.power(ux, q),x[:,0]*x[:,1]))
    sum2 = np.sum(np.multiply(np.power(ux,q), np.square(x[:,0]) - np.square(x[:,1])))
    return 0.5*math.atan2(2*sum1,sum2)
```

`cost_function` determina el valor de la función de coste en un instante determinado

```python
def cost_function (x, ux, theta_cluster, q = 1): #q = 1 for crisp clustering
    
    # x = conjunto de puntos 
    # ux = pertenencias
    # theta_cluster = parámetros del conjunto de clusters 
    # q = fuzzyfier
    
    cost = 0
    for i, theta_clus in enumerate(theta_cluster):
        sum_ = np.sum(np.power(ux[:,i],q) * distance_array(x,theta_clus))
        cost += sum_
    return cost
```

`do_crisp_clust` implementa el pseudocódigo de la diapositiva 7.

```python
def do_crisp_clust(X, M, n_iter, n_attempts, eps):
    cost_clustering = np.zeros((n_iter, n_attempts)) 
    u_clustering = np.zeros((n_attempts, X.shape[0], M))
    theta_clustering = np.zeros((n_attempts, M))
    last_value_cost = np.zeros((n_attempts,))
    
    for attempt in range(n_attempts):
        theta = np.random.uniform(low = 0, high = 2*np.pi, size = (M,)) #inicialización aleatoria de clusters
        theta_p = np.zeros(theta.shape)
        iters = 0
        prev = 1e03
        while iters < n_iter: 
            u = np.zeros((X.shape[0], M)) #pertenencias a cluster -> {0,1}
            if iters > 0:
                theta = theta_p.copy()
            for i,point in enumerate(X): 
                u_aux = np.zeros((M,))
                for j,cluster in enumerate(theta): 
                    u_aux[j] = distance(point, cluster) #distancia entre punto y cluster
                u[i,np.argmin(u_aux)] = 1 # se asigna el cluster más cercano

            for k in range(M):
                theta_p[k] = theta_clust(X, u[:,k]) #nuevos parámetros theta
                
            cost_clustering[iters, attempt] = cost_function(X, u, theta_p)
            # si J(t) - J(t-1) < eps
            if np.abs(cost_clustering[iters, attempt] - prev) < eps:
                last_value_cost[attempt] = cost_clustering[iters, attempt]
                break
            else: 
                prev = cost_clustering[iters, attempt]
            iters += 1
            
        u_clustering[attempt] = u
        theta_clustering[attempt, :] = theta_p
    
    optimum_clustering = np.argmin(last_value_cost) #valor más bajo del coste en la última iteración
    
    # pertencias, valores del coste del mejor clustering, parámetros del mejor clustering
    return u_clustering[optimum_clustering], cost_clustering[:, optimum_clustering], theta_clustering[optimum_clustering, :] 
```

`fuzzy_membership` determina la pertenencia borrosa $u$ de un punto a un cluster.

```python
def fuzzy_membership (point, dist, theta_clus, q):
    
    #point = observación evaluada
    # dist = distancia entre el punto y el cluster evaluado
    #theta_clus = parámetros del conjunto de clusters
    # q = fuzzyfier
    
    total_sum = 0
    for theta in theta_clus: 
        sum_ = (dist/distance(point, theta))**(1/(q-1))
        total_sum += sum_
    return 1/total_sum
```

```python
def do_fuzzy_clust(X, M, n_iter, n_attempts, eps, q):
    cost_clustering = np.zeros((n_iter, n_attempts))
    u_clustering = np.zeros((n_attempts, X.shape[0], M))
    theta_clustering = np.zeros((n_attempts, M))
    last_value_cost = np.zeros((n_attempts,))
    
    for attempt in range(n_attempts):
        theta = np.random.uniform(low = 0, high = 2*np.pi, size = (M,)) #inicialización aleatoria de clusters
        theta_p = np.zeros(theta.shape)
        iters = 0
        prev = 1e03
        while iters < n_iter: 
            u = np.zeros((X.shape[0], M)) #pertenencias a cluster -> {0,1}
            if iters > 0:
                theta = theta_p.copy()
                
            for i,point in enumerate(X): 
                u_aux = np.zeros((M,))
                for j,cluster in enumerate(theta): 
                    u[i,j] = fuzzy_membership(point, distance(point, cluster), theta, q) #pertenencia a cada cluster

            for k in range(M):
                theta_p[k] = theta_clust(X, u[:,k], q)
                
            cost_clustering[iters, attempt] = cost_function(X, u, theta_p)
            if np.abs(cost_clustering[iters, attempt] - prev) < eps:
                last_value_cost[attempt] = cost_clustering[iters, attempt]
                break
            else: 
                prev = cost_clustering[iters, attempt]
            iters += 1
            
        u_clustering[attempt] = u
        theta_clustering[attempt, :] = theta_p
    
    optimum_clustering = np.argmin(last_value_cost) #valor más bajo del coste en la última iteración
    
    # pertencias, valores del coste del mejor clustering, parámetros del mejor clustering
    return u_clustering[optimum_clustering], cost_clustering[:, optimum_clustering], theta_clustering[optimum_clustering, :] 
```

Al tratarse de un caso donde la clasificación no es discreta o binaria, sino con valores difusos, no se puede establecer una relación directa entre etiquetas reales (crisp) y estimadas (fuzzy). Por lo tanto, no es posible el cálculo de la tabla de contingencia o de la métrica V-measure, o al menos con lo conocimiento que se tienen. 

Una posibilidad de estimación tanto de la tabla como de la V-measure sería redondear la etiqueta de la clase al valor fuzzy más grande. Sin embargo, se desconoce si esto sería o no una buena práctica, ya que entonces se pierde la esencia de este tipo de métodos difusos. 

## DBSCAN

```python
from sklearn.cluster import DBSCAN
from sklearn.model_selection import ParameterGrid
```

```python
group = '02' 
ds = 4
data = np.loadtxt(path + 'ds'+group+str(ds)+'.txt')

X4 = data[:, 0:2]
y4 = data[:, 2:3].ravel()
```

`evaluate_DBSCAN` perimitirá realizar un Grid Search con los parámetros indicados en `comb`. 

```python
def evaluate_DBSCAN (X,y,comb): 
    
    # X = conjunto de datos
    # y = etiquetas 
    # comb = conjunto de parámetros
    
    iters = len(comb)
    labeling = np.zeros((X.shape[0], iters)) #n_samples x models
    v_measure = np.zeros((iters,))
    for i in range (iters): 
        dbscan = DBSCAN(eps = comb[i]['eps'], min_samples = comb[i]['min_samples']) #inicialización de DBSCAN con parámetros requeridos
        pred_labels = dbscan.fit_predict(X) #clustering
        labeling[:,i] = pred_labels
        v_measure[i] = metrics.v_measure_score(y, pred_labels) #V-measure
        
    # clasificación de las muestras, V-measure para cada conjunto de parámetros
    return labeling, v_measure
```

```python
values = {'eps': [0.1, 0.2, 0.3, 0.4], 
          'min_samples' : [3,5]} #parámetros a estudiar

param_grid = list(ParameterGrid(values)) #se crea un grid de parámetros sobre el que realizar el Grid Search
labels, v = evaluate_DBSCAN (X4, y4, param_grid)
```

```python
print('Best performing values: ')
print('Eps = ', param_grid[np.argmax(v)]['eps'])
print('Min_samples = ', param_grid[np.argmax(v)]['min_samples'])
```

```python
# se imprimen los valores de V-measure por cada conjunto empleado
for n, conjunto in enumerate(param_grid):
    print('eps = {e}, min_samples = {m} -> v_measure = {vm}'.format(e = conjunto['eps'], m = conjunto['min_samples'], vm = v[n]))
```
